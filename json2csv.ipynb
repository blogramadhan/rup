```python
dfp = pd.read_csv("data/pp.csv")
```

### **RUP Penyedia - Metode Pengadaan - Persentase**

```python
mp_persen = dfp['metodepengadaan'].value_counts(normalize=True)
mp_persen
```

```python
plot_mp_persen = mp_persen.plot.pie(subplots=True, figsize=(16, 8))
plot_mp_persen
```

### **RUP Penyedia - Metode Pengadaan - Nilai Pagu**

```python
mp_nilai = dfp.groupby('metodepengadaan')['jumlahpagu'].sum()
mp_nilai
```

```python
plot_mp_nilai = mp_nilai.plot.pie(subplots=True, figsize=(16, 8))
plot_mp_nilai
```

### **RUP Penyedia - Jenis Pengadaan - Persentase**

```python
jp_persen = dfp['jenispengadaan'].value_counts(normalize=True)
jp_persen
```

```python
plot_jp_persen = jp_persen.plot.pie(subplots=True, figsize=(16, 8))
plot_jp_persen
```

### **RUP Penyedia - Jenis Pengadaan - Nilai Pagu**
